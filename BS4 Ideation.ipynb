```python
# Printing the first match

req1 = requests.get(season_url[0])
print(req1)
resp1 = HtmlResponse('example.com', body=req1.text, encoding= 'utf-8')

first=resp1.xpath("//div[@class='ds-flex']//a[@class='ds-no-tap-higlight']/@href")

print(first.getall())
```

```python
# Looping through the season_url to pull individual match urls from all seasons

all_match_url=[]

for url in season_url:
    req=requests.get(url)
    resp = HtmlResponse('example.com', body=req.text, encoding= 'utf-8')
    
    match_link = resp.xpath("//div[@class='ds-flex']//a[@class='ds-no-tap-higlight']/@href")
    all_match_url.append(match_link.getall())
```

```python
# Unpacking the list of lists, so that we simplify futher work

unpack_all_match_url = [element for sublist in all_match_url for element in sublist]
print(upack_all_match_url)
print(len(unpack_all_match_url))
```

```python
# Creating a new list to concatenate the domain with the individual match urls

match_url_all =[crickinfo + url for url in unpack_all_match_url]

print(len(match_url_all))
```

```python
# SEVENTH

points=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr//span[@class='ds-text-tight-s ds-font-regular'])[4]/text()")
points.get()
```

```python
# Finally creating a loop to get the meta-data as above and some more from each 1032 matches, the entire history of IPL
# No need to run, as this is a long loop and might run as long as 25-30 min!


stadium_list=[]
team_a_list =[]
team_b_list=[]
toss_list = []
player_of_match_list = []
hours_of_play_list = []
date_of_match_list = []
debut_player_list = []
umpire_1_list = []
umpire_2_list = []
tv_umpire_list = []
reserve_umpire_list = []
match_refree_list = []
points_list = []

for match in match_url_all:
    req=requests.get(match)
    resp = HtmlResponse('example.com', body=req.text, encoding= 'utf-8')
    
    # 01
    stadium = resp.xpath("//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max']/a/@title")
    stadium_list.append(stadium.getall())

    # 02
    team_a = resp.xpath("(//div[@class='ds-rounded-lg ds-mt-2']//span/span/text())[1]")
    team_a_list.append(team_a.getall())

    # 03
    team_b = resp.xpath("((//div[@class='ds-rounded-lg ds-mt-2'])[2]//span/span/text())[1]")
    team_b_list.append(team_b.getall())
    
    # 04
    toss = resp.xpath("//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//td[@class='ds-min-w-max ds-text-typo']/span/text()")
    toss_list.append(toss.get())
    
    # 05
    player_of_match=resp.xpath("//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//td[@class='ds-min-w-max ds-text-typo']//a[@class='ds-flex ds-items-center ds-cursor-pointer']/span/@title")
    player_of_match_list.append(player_of_match.getall())
    
    # 06 
    hours_of_play=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr//td[@class='ds-min-w-max ds-text-typo']/span[@class='ds-text-tight-s ds-font-regular'])[2]/text()")
    hours_of_play_list.append(hours_of_play.getall())
    
    # 07
    date_of_match = resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr//td[@class='ds-min-w-max ds-text-typo']/span[@class='ds-text-tight-s ds-font-regular'])[3]/text()")
    date_of_match_list.append(date_of_match.getall())

    # 08
    debut_player=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max ds-text-typo']//div//span[@class='ds-cursor-pointer ds-inline-flex ds-items-start ds-leading-none']/@title)[2]")
    debut_player_list.append(debut_player.getall())

    # 09
    umpire_1=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max ds-text-typo !ds-p-0']//div//span[@class='ds-cursor-pointer ds-inline-flex ds-items-start ds-leading-none']/@title)[1]")
    umpire_1_list.append(umpire_1.getall())

    umpire_2=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max ds-text-typo !ds-p-0']//div//span[@class='ds-cursor-pointer ds-inline-flex ds-items-start ds-leading-none']/@title)[2]")
    umpire_2_list.append(umpire_2.getall())

    # 10
    tv_umpire=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max ds-text-typo']//div//span[@class='ds-cursor-pointer ds-inline-flex ds-items-start ds-leading-none']/@title)[3]")
    tv_umpire_list.append(tv_umpire.getall())

    # 11
    reserve_umpire=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max ds-text-typo']//div//span[@class='ds-cursor-pointer ds-inline-flex ds-items-start ds-leading-none']/@title)[4]")
    reserve_umpire_list.append(reserve_umpire.getall())

    # 12
    match_refree=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr/td[@class='ds-min-w-max ds-text-typo']//div//span[@class='ds-cursor-pointer ds-inline-flex ds-items-start ds-leading-none']/@title)[5]")
    match_refree_list.append(match_refree.getall())

    # 13
    points=resp.xpath("(//table[@class='ds-w-full ds-table ds-table-sm ds-table-auto ']//tr//span[@class='ds-text-tight-s ds-font-regular'])[4]/text()")
    points_list.append(points.getall())
```

```python
# Printing the lengths of each list individually so that the length is verified

print(
    f"Stadiums: {len(stadium_list)}\n"
    f"Team_A: {len(team_a_list)}\n"
    f"Team_B: {len(team_b_list)}\n"
    f"Tosses: {len(toss_list)}\n"
    f"Players of Match: {len(player_of_match_list)}\n"
    f"Hours of Play: {len(hours_of_play_list)}\n"
    f"Dates of Match: {len(date_of_match_list)}\n"
    f"Debut Players: {len(debut_player_list)}\n"
    f"Umpire 1s: {len(umpire_1_list)}\n"
    f"Umpire 2s: {len(umpire_2_list)}\n"
    f"TV Umpires: {len(tv_umpire_list)}\n"
    f"Reserve Umpires: {len(reserve_umpire_list)}\n"
    f"Match Referees: {len(match_refree_list)}\n"
    f"Points: {len(points_list)}"
)
```

```python
# Creating a schema to properly showcase the meta-data of all matches

match_data = {
    "Stadiums": stadium_list,
    "Team_A": team_a_list,
    "Team_B": team_b_list,
    "Tosses": toss_list,
    "Players of Match": player_of_match_list,
    "Hours of Play": hours_of_play_list,
    "Dates of Match": date_of_match_list,
    "Debut Players": debut_player_list,
    "Umpire 1s": umpire_1_list,
    "Umpire 2s": umpire_2_list,
    "TV Umpires": tv_umpire_list,
    "Reserve Umpires": reserve_umpire_list,
    "Match Referees": match_refree_list,
    "Points": points_list
}

print(match_data)
```
